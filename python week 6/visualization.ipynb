```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = 'gym_members_exercise_tracking.csv'

try:
    data = pd.read_csv(file_path)
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: File not found. Please check the file path.")

# Display the first few rows
print("First few rows of the dataset:")
print(data.head())

# Check the data structure and types
print("\nDataset Info:")
print(data.info())

# Check for missing values
print("\nMissing Values:")
print(data.isnull().sum())

# Clean missing values (if any)
data.fillna(data.mean(), inplace=True)
print("Missing values have been filled.")
```

```python
# Basic statistics for numerical columns
print("\nBasic Statistics:")
print(data.describe())

# Grouping and computing mean for a numerical column (e.g., Calories_Burned) by a categorical column (e.g., Workout_Type)
avg_calories_by_workout = data.groupby('Workout_Type')['Calories_Burned'].mean()
print("\nAverage Calories Burned by Workout Type:")
print(avg_calories_by_workout)

# Grouping by Gender and computing average BMI
avg_bmi_by_gender = data.groupby('Gender')['BMI'].mean()
print("\nAverage BMI by Gender:")
print(avg_bmi_by_gender)
```

```python
# 1. Line Chart: Average BPM by Experience Level
plt.figure(figsize=(10, 6))
sns.lineplot(x="Experience_Level", y="Avg_BPM", data=data, marker="o", errorbar=None)
plt.title("Average BPM by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average BPM")
plt.grid(True)
plt.show()

# 2. Bar Chart: Average Calories Burned by Workout Type
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_calories_by_workout.index, y=avg_calories_by_workout.values, palette="viridis")
plt.title("Average Calories Burned by Workout Type")
plt.xlabel("Workout Type")
plt.ylabel("Calories Burned")
plt.show()

# 3. Histogram: Distribution of BMI
plt.figure(figsize=(10, 6))
plt.hist(data['BMI'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()

# 4. Scatter Plot: Relationship Between Weight and Calories Burned
plt.figure(figsize=(10, 6))
plt.scatter(data['Weight (kg)'], data['Calories_Burned'], alpha=0.7, color='teal')
plt.title("Weight vs. Calories Burned")
plt.xlabel("Weight (kg)")
plt.ylabel("Calories Burned")
plt.grid(True)
plt.show()
```
